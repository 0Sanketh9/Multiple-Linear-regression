```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.formula.api as smf
import statsmodels.api as sm
from statsmodels.graphics.regressionplots import influence_plot
```

```python
startup=pd.read_csv("50_Startups.csv")
startup
```

```python
startup.info()
```

```python
startup1=startup.rename({'R&D Spend':'RDS','Administration':'ADM','Marketing Spend':'MKS'},axis=1)
startup1
```

```python
startup1.describe()
```

```python
#correlation analysis
startup1.corr()
```

```python
sns.set_style(style='darkgrid')
sns.pairplot(startup1)
```

```python
model=smf.ols("Profit~RDS+ADM+MKS",data=startup1).fit()
```

```python
model.summary()
```

```python
#building SLR and MLR for insignificant terms like "ADM"and "MKS" and finding t,p values
```

```python
slr_a=smf.ols("Profit~ADM",data=startup1).fit()
slr_a.tvalues , slr_a.pvalues                    #From pvalue ADM is insignificant 
```

```python
slr_a=smf.ols("Profit~MKS",data=startup1).fit()
slr_a.tvalues , slr_a.pvalues                   #From pvalue MKS is significant
```

```python
mlr_am=smf.ols("Profit~ADM+MKS",data=startup1).fit()
mlr_am.tvalues , mlr_am.pvalues               #both has significant p values
```

```python
#Calculating VIF
rsq_r=smf.ols("RDS~ADM+MKS",data=startup1).fit().rsquared
vif_r=1/(1-rsq_r)

rsq_a=smf.ols("ADM~RDS+MKS",data=startup1).fit().rsquared
vif_a=1/(1-rsq_a)

rsq_m=smf.ols("MKS~RDS+ADM",data=startup1).fit().rsquared
vif_m=1/(1-rsq_m)

# Putting the values in Dataframe format
d1={'Variables':['RDS','ADM','MKS'],'Vif':[vif_r,vif_a,vif_m]}
Vif_df=pd.DataFrame(d1)
Vif_df
```

```python
#Residual Analysis
sm.qqplot(model.resid,line='q')
plt.title("Normal Q-Q plot of residuals")
plt.show()
```

```python
list(np.where(model.resid<-30000))
```

```python
#Residual plot for Homoscedasticity
def standard_values(vals) : return (vals-vals.mean())/vals.std()  
```

```python
plt.scatter(standard_values(model.fittedvalues),standard_values(model.resid))
plt.title('Residual Plot')
plt.xlabel('standardized fitted values')
plt.ylabel('standardized residual values')
plt.show() 
```

```python
#Residual V/s Regressors
fig=plt.figure(figsize=(15,8))
sm.graphics.plot_regress_exog(model,'RDS',fig=fig)
plt.show()
```

```python
fig=plt.figure(figsize=(15,8))
sm.graphics.plot_regress_exog(model,'ADM',fig=fig)
plt.show()
```

```python
fig=plt.figure(figsize=(15,8))
sm.graphics.plot_regress_exog(model,'MKS',fig=fig)
plt.show()
```

```python
#Model Deletion Diagnostic,detecting outliers/influencers: cook's distance and leverage value
```

```python
model_influence=model.get_influence()
(c,_)=model_influence.cooks_distance
```

```python
fig=plt.figure(figsize=(20,7))
plt.stem(np.arange(len(startup1)),np.round(c,5))
plt.xlabel('Row Index')
plt.ylabel('Cooks Distance')
plt.show()
```

```python
# Index and value of influencer where C>0.5
np.argmax(c) , np.max(c)
```

```python
# 2. Leverage Value using High Influence Points : Points beyond Leverage_cutoff value are influencers
influence_plot(model)
plt.show()
```

```python
# Leverage Cuttoff Value = 3*(k+1)/n ; k = no.of features/columns & n = no. of datapoints
k=startup1.shape[1]
n=startup1.shape[0]
leverage_cutoff = (3*(k+1))/n
leverage_cutoff
```

```python
startup1[startup1.index.isin([49])] 
```

```python
#Improving the model
#Discard the data points which are influencers and reassign the row number (reset_index(drop=True))
startup2=startup1.drop(startup1.index[[49]],axis=0).reset_index(drop=True)
startup2
```

```python
#BUILD Model
```

```python
final_ml_V=smf.ols("Profit~RDS+MKS",data=startup2).fit()
```

```python
(final_ml_V.rsquared,final_ml_V.aic,final_ml_V.rsquared_adj)
```

```python
#Model dletion diagnostics and Final model
while np.max(c)>0.5 :
    model=smf.ols("Profit~RDS+ADM+MKS",data=startup2).fit()
    (c,_)=model.get_influence().cooks_distance
    c
    np.argmax(c) , np.max(c)
    startup2=startup2.drop(startup2.index[[np.argmax(c)]],axis=0).reset_index(drop=True)
    startup2
else:
    final_model=smf.ols("Profit~RDS+ADM+MKS",data=startup2).fit()
    final_model.rsquared , final_model.aic
    print("Thus model accuracy is improved to",final_model.rsquared)
```

```python
# Manual Prediction of Price
final_model.predict(startup2)
```
